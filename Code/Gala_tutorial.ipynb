```python
import astropy.units as u
import numpy as np
import gala.potential as gp
import pandas as pd
from astropy.table import Column, join, Table, vstack, hstack
from astropy.io import ascii
```

```python
HMXB_table = ascii.read('HMXB_20241202_.csv',format='csv')
```

```python
mw = gp.MilkyWayPotential()
```

```python
mw
```

```python
xyz = [-8.,0,0] * u.kpc
mw.energy(xyz)
```

```python
import gala.dynamics as gd
```

```python
w0 = gd.PhaseSpacePosition(pos=[-8.1,0,0.02]*u.kpc, vel=[13, 245,8.]* u.km/u.s)
```

```python
orbit = mw.integrate_orbit(w0, dt= 1*u.Myr, t1=0, t2=2*u.Gyr)
```

```python
orbit.plot(['x','y'])
```
